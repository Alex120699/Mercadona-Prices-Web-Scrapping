```python
from operator import index


antes = pd.read_csv(f"mercadona_prices_{codigo_postal}_2022-08-11.csv",index_col=0)
despues = pd.read_csv(f"mercadona_prices_{codigo_postal}_2022-10-01.csv",index_col=0)
```

```python
eliminados, variaciones_precios = [],[]
ids_siguen = []

for prod,cant,prec in zip(antes.Producto,antes.Cantidad,antes.Precio):
    #print(prod,"/",cant,"/",prec)
    df_comun = despues[(despues.Producto==prod) & (despues.Cantidad==cant)]
    for id in df_comun.index:
        ids_siguen.append(id)

    if df_comun.shape[0]==0:
        print("\n--------------------------------------------\n",prod,cant,"\n")
        print("Lo han eliminado o ha cambiado la cantidad","\n--------------------------------------------\n")
        eliminados.append([prod,cant,prec])
    elif df_comun.shape[0]>1:
        print("\n--------------------------------------------\n",prod,cant,"\n")
        print("Está repetido el producto","\n--------------------------------------------\n")
    else:
        row_ahora = df_comun.iloc[0,:]
        #print(row_ahora.Producto,"/",row_ahora.Cantidad,"/",row_ahora.Precio)
        precio_ahora = (despues[(despues.Producto==prod) & (despues.Cantidad==cant)]).Precio.iloc[0]

        if (float(prec)-precio_ahora)!=0:
            print("\n--------------------------------------------\n",prod,cant,"\n")
            print("PRECIO ANTES: ",prec,"\nPRECIO AHORA: ",precio_ahora,"\n--------------------------------------------\n")
            variaciones_precios.append([prod,cant,prec,precio_ahora])

eliminados_df = pd.DataFrame(eliminados,columns=["Producto","Cantidad","Precio"])
nuevos_df = despues.drop(np.unique(ids_siguen),axis=0)[["Producto","Cantidad","Precio"]]
variaciones_precios_df = pd.DataFrame(variaciones_precios,columns=["Producto","Cantidad","Precio_antes","Precio_ahora"])
```
